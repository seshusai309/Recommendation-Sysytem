```python
# importing Libraries
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# to ignore warnings 
import warnings
warnings.filterwarnings('ignore')
```

```python
# import the three data files from the folder of csv files
books = pd.read_csv('recommendation/data/Books.csv')
ratings = pd.read_csv('recommendation/data/Ratings.csv')
users = pd.read_csv('recommendation/data/Users.csv')
```

```python
# delete the rows where there is null values
books = books.dropna()
```

```python
users = users.dropna()
```

```python
np.sort(ratings["Book-Rating"].unique())
```

```python
books['Year-Of-Publication'] = books['Year-Of-Publication'].astype('int32')
```

```python
books_with_ratings = ratings.merge(books, on='ISBN')
```

```python
popular_df = books_with_ratings.groupby('Book-Title').agg(num_rating=('Book-Rating', 'count'), 
                                                         avg_rating=('Book-Rating', 'mean'))
```

```python
popular_df = popular_df.reset_index()
```

```python
popular_df.sort_values('num_rating', ascending=False)
```

```python
# popularity will always depends on the number of people that read
# Based on Rating 
popular_df = popular_df[popular_df['num_rating']>300].sort_values('avg_rating', ascending=False)
```

```python
popular_df = popular_df.head(50)
```

```python
# For the Model Deployment I need Book_title, Author, Image_Url
popular_df = popular_df.merge(books, on = "Book-Title").drop_duplicates('Book-Title')[['Book-Title',
                                                                                       'Book-Author',
                                                                                       'Image-URL-M',
                                                                                      'num_rating',
                                                                                      'avg_rating']]
```

```python
popular_df
```

```python
x = books_with_ratings.groupby('User-ID').count()
```

```python
x = x['Book-Rating'] > 200
```

```python
power_users = x[x].index
```

```python
# selecting only the records of power_users
filterd_ratings = books_with_ratings[books_with_ratings['User-ID'].isin(power_users)]
```

```python
y = filterd_ratings.groupby('Book-Title').count()
```

```python
y.sort_values('User-ID', ascending=False)
```

```python
y = y['User-ID'] >= 50
y
```

```python
famous_books = y[y].index
```

```python
final_ratings = filterd_ratings[filterd_ratings['Book-Title'].isin(famous_books)]
final_ratings
```

# Pivot Table giving the ratings for each book from each User
#Book row With usserId as Column

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
pt
```

```python
pt = pt.fillna(0)
```

```python
similarity_scores = cosine_similarity(pt)
similarity_scores
```

```python
df_temp = pd.DataFrame(similarity_scores)
df_temp
```

```python
df_temp_name = df_temp
```

```python
df_temp_name.index = pt.index
df_temp_name.columns = pt.index
```

```python
pt.index
```

```python
def recommand(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x : x[1], reverse=True)[1:6]
    data = []
    for i in similar_items:
        item = [] 
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data
    
```

```python
recommand('2nd Chance')
```
